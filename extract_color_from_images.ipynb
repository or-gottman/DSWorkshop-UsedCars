### Download images

```python
import urllib.request
import pathlib
import os

download_dir = './data/images/250k/'
pathlib.Path(download_dir).mkdir(parents=True, exist_ok=True)

successful_downloads, failed_downloads = 0, 0
limit = -1

for row in imgs_df.itertuples():
    image_url = row.main_picture_url
    car_id = str(row.id)
    img_extension = image_url.split('.')[-1]
    img_save_name = f'{car_id}.{img_extension}'
    if os.path.isfile(download_dir + img_save_name):
        print('skipping', img_save_name)
        continue

    print(img_save_name)
    try:
        urllib.request.urlretrieve(image_url, download_dir + img_save_name)
        successful_downloads += 1
    except Exception as e:
        print("Error on url:", image_url)
        print(e.args)
        failed_downloads += 1

    limit -= 1
    if limit == 0:
        break
print("Images downloaded:", successful_downloads)
print("Images failures:", failed_downloads)
```

## Remove images backgrounds to better determine the cars' colors

```python
import os
from images_utils import remove_background

imgs_without_bg_dir = './data/images/250k_no_bg/'
pathlib.Path(imgs_without_bg_dir).mkdir(parents=True, exist_ok=True)

for image in os.scandir(download_dir):
    if image.is_file() and not os.path.isfile(imgs_without_bg_dir + image.name):
        try:
            remove_background(image.path, imgs_without_bg_dir + image.name.split('.')[0] + '.png')
            print(image.name)
        except Exception as e:
            pass
    # else:
    #     print('skipping', image.name)
```

## Find dominant color of images

```python
imgs_df
```

```python
# from images_utils import find_dominant_color_v1 as find_dominant_color
from images_utils import find_dominant_color_v2 as find_dominant_color
from images_utils import convert_rgb_to_names
from PIL import Image

for row in imgs_df.itertuples():
    car_id = str(row.id)
    # print(imgs_without_bg_dir + car_id)
    if os.path.exists(imgs_without_bg_dir + car_id + '.png'):
        img = Image.open(imgs_without_bg_dir + car_id + '.png')
        try:
            dominant_color = find_dominant_color(img)
            color_name = convert_rgb_to_names(dominant_color)

            # Generalize colors (as some colors are extracted as lightgray, rosebrown and similar)
            for general_color in ['blue', 'red', 'black', 'gray', 'grey', 'orange', 'white', 'green', 'yellow', 'brown']:
                if general_color in color_name:
                    if general_color == 'grey':
                        general_color = 'gray'
                    color_name = general_color

        except Exception as e:
            continue

        imgs_df.at[row.Index, 'listing_color'] = color_name.upper()
```
